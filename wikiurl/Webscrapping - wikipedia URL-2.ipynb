```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_national_parks_of_the_United_States'
```

```python
page = requests.get(url)
```

```python
page
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
soup.head
```

```python
soup.find_all('table')[1]
```

```python
soup.find('table', class_ = 'wikitable sortable')
```

```python
table = soup.find_all('table')[1]
```

```python
table
```

```python
#get the headers form the table whcih is nothing but the <th> tags
titles = soup.find_all('th')
```

```python
titles
```

```python
table_titles = [ title.text.strip() for title in titles]
print(table_titles)
```

```python
table_titles_header = table_titles[0:7]
table_titles_header
```

```python
table_titles_name = table_titles[7:-6]
```

```python
table_titles_name
```

```python
import pandas as pd
df = pd.DataFrame(columns = table_titles_header)
```

```python
df
```

```python
#column data with <tr> rows , <td> with data tags
column_data = table.find_all('tr')
```

```python
column_data
```

```python
for row in column_data[1:]: 
    row_data = row.find_all('td')
    individual_row= [data.text.strip() for data in row_data]
    print(individual_row)
    #length = len(df)
    #df.loc[length] = individual_row
```
